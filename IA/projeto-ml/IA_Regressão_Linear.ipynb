# Jupyter Notebook Para desenvolvimento de uma IA que prevê preços.

Ela utiliza uma base de dados para prevê atravês de Regressão Linear o preço da pizza atravês do diametro.

O projeto roda na web para melhor visualização.

---

Primeiro armazenamos em uma variavel o arquivo que contém nossos dados.

```python
import pandas as pd

df = pd.read_csv("pizzas.csv")
```

## Criamos uma vizualização em gráfico dos nossos dados.

```python
df.plot(kind="scatter", x="diametro", y="preco");
```

Criamos um modelo "modelo" que armazena o metodo de **Regressão Linear** do objeto **"Modelo Linear"**.

Armazenamos os **valores** das colunas X e Y em variaveis.

Por fim, treinamos esse modelo com as colunas.

```python
from sklearn import linear_model

modelo = linear_model.LinearRegression()
x = df[["diametro"]]
y = df[["preco"]]

modelo.fit(x, y)
```

### E imprimimos a predição desse Modelo tendo como valor Y

```python
Y = int(input("Qual o diametro da pizza? "))

print(modelo.predict([[Y]])[0][0])
```
